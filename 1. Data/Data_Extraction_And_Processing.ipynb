```python
#Calculating the Goals For (GF), Goals Against, and Win-Draw-Loss Record for each team

y=sb.matches(competition_id=2, season_id=27)
y=y[["home_score", "away_score", "home_team", "away_team"]]
```

```python
gf=[]
ga=[]
pts=[]
w=[]
d=[]
l=[]
```

```python
#Creating a DataFrame for the final PL Table
table=pd.DataFrame()
table['team']=y.home_team.unique()
```

```python
for i in table.team:
    wins=0
    draws=0
    c=0
    a=y[y.home_team==i]
    hs=sum(a.home_score)
    hc=sum(a.away_score)
    for f in a.reset_index().index:
        if a.home_score.iloc[f]>a.away_score.iloc[f]:
            wins+=1
            c+=3
        elif a.home_score.iloc[f]==a.away_score.iloc[f]:
            draws+=1
            c+=1
    b=y[y.away_team==i]
    aws=sum(b.away_score)
    awc=sum(b.home_score)
    for f in b.reset_index().index:
        if b.away_score.iloc[f]>b.home_score.iloc[f]:
            wins+=1
            c+=3
        elif b.away_score.iloc[f]==b.home_score.iloc[f]:
            draws+=1
            c+=1
    gf.append((hs+aws))
    ga.append(hc+awc)
    w.append(wins)
    d.append(draws)
    pts.append(c)
    
```

```python
table['gf']=gf
table['ga']=ga
table['gd']=table.gf-table.ga
table['w']=w
table['d']=d
table['l']=0
table['points']=pts
```

```python
table.l=38-table.w-table.d
```

```python
table=table.sort_values(by='points', ascending=False).reset_index().drop(columns='index')
```

```python
shots=grouped_events["shots"]
dribbles=grouped_events['dribbles']
interceptions=grouped_events['interceptions']
fouls=grouped_events['foul_committeds']
errors=grouped_events['errors']
bad=grouped_events['bad_behaviours']
```

```python
metrics=[shots,dribbles,interceptions,fouls,errors,bad]
mt=['shots','dribbles','interceptions','fouls','errors','bad_behaviour']
```

```python
for f in range(len(metrics)):
    i=metrics[f]
    x=i[i.possession_team.isin(top_six)].groupby('possession_team').agg({'id':'count'}).reset_index()
    x=x.rename(columns={'possession_team':'team','id': mt[f]})
    t6=pd.merge(on='team', left=t6, right=x, how='inner')
```

```python
for i in t6.columns:
    if i in mt:
        t6[i]=round(t6[i]/38,2)
```
